```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = pd.read_csv('no_dup_data.csv')
```

```python
df.rename({"request_type_title" : "issue_type", 
           "web_url" : "url"}, axis=1,inplace=True)
```

```python
df.columns
```

```python
df["created_at"]
```

```python
dateNew = []
for i in range(318592):
    dto_str = df["created_at"][i][0:10]
    dto = datetime.strptime(dto_str, '%Y/%m/%d')
    dateNew.append(dto)
```

```python
dateNew[0]
```

```python
df["date"] = dateNew
df["date"].head()
```

```python
av = df[df["issue_type"] == "Abandoned Vehicle"]
avDate = av[["date", "issue_type"]]
avDate = avDate.reset_index()
avDate = avDate[["date", "issue_type"]]

avIndex = [i for i in range(24707)]
#avDate["index"] = avIndex
avCount = avDate.groupby(["date"]).count()
avCount.rename({"issue_type" : "frequency"}, axis=1,inplace=True)
#avFreq.count()["index"]
avCount.count()
```

```python
avCount.plot(figsize=(10, 10))
```

```python
av1 = av[["date"]]
avCreate = [pd.to_datetime(av["created_at"])]
            
# av.created_at = pd.to_datetime(av.created_at)
av1 = av1.set_index(avCreate)
av1.head()
av1 = av1.resample('1D').count()
avCount.plot(figsize=(10, 10))
```
